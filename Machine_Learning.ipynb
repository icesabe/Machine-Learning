## Get features

```python
def processing(df, sheet_name, drop_columns=None, head_cat=None):
    df_tmp = pd.read_excel(df, sheet_name)
    if drop_columns is not None:
        df_tmp.drop(columns=drop_columns, inplace=True)        
        
    # print_head(df_tmp)
    df_tmp.replace({float('nan'): 0.0, '-': 'NA'}, inplace=True) ###missing data
    
    # drop empty columns 
    cols = [] 
    for col in df_tmp.columns: 
        cols.append(col)
        
    for col in cols: 
        if 'Unnamed' in col: 
            df_tmp.drop(columns=col, inplace=True)
        
    if head_cat is not None:
        for val in print_head(df_tmp): 
            if str(val) in head_cat: 
                df_tmp[str(val)] = df_tmp[str(val)].astype('category')
                
    # dealing with nan, special characters(-), and categorical values. TODO
    
    return df_tmp
            
```

```python
'''
data processing:
1. identify the missing value index
2. remove the corresponding data sample (feature and label)
'''
def eval_metric(model, test_x, test_y):
    print(sum(test_y), sum(1-test_y))
    pred_y = model.predict(test_x)
    print(confusion_matrix(test_y, pred_y))
    
    tn, fp, fn, tp = confusion_matrix(test_y, pred_y).ravel()    # return an ndarray
    print(f'true neg, false pos, false neg, true pos:')
    print(tn, fp, fn, tp)
    print("Accuracy:", float(tn + tp) / sum([tn, fp, fn, tp]))
    print("F1 scores:", f1_score(test_y, pred_y))     # average='weighted' / zero_division
    print("Report:", classification_report(test_y, pred_y))
    
def eval_mix(model, test_x, test_y):
    print(sum(test_y), sum(1-test_y))
    pred_mix_y = model.predict(test_x)
    pred_mix_y = pred_mix_y.flatten()
    y_pred = np.where(pred_mix_y > 0.5, 1, 0)
    print(confusion_matrix(test_y, y_pred))
    
    tn, fp, fn, tp = confusion_matrix(test_y, y_pred).ravel()    # return an ndarray
    print(f'true neg, false pos, false neg, true pos:')
    print(tn, fp, fn, tp)
    print("Accuracy:", float(tn + tp) / sum([tn, fp, fn, tp]))
    print("F1 scores:", f1_score(test_y, y_pred))     # average='weighted' / zero_division
    print("Report:", classification_report(test_y, y_pred))
    

def main_process(name='高脂血症', X=None, seed=2022):
    if X is None:
        print('no feature provided')
        return -1
    # data processing:
    #Y = label_values[name]
    Y = name

    indexes = np.argwhere(np.isnan(Y))

    print(f'=== Indexes of samples with missing label: {indexes}')

    Y_f = np.delete(np.array(Y), indexes, axis=0)
    X_f = np.delete(X, indexes, axis=0)

    print('=== Result data shape', Y_f.shape, X_f.shape)

    print(f"=== Total number of sampel: {Y_f.shape[0]}, pos: {sum(Y_f)}, neg: {sum(1 - Y_f)}, pos_rate: {sum(Y_f)/float(Y_f.shape[0])}")
    
    # data split
    np.random.seed(seed)
    all_index = np.arange(len(Y_f))
    np.random.shuffle(all_index)

    n_train = int(len(all_index) * 0.7)

    train_index = all_index[0: n_train]
    test_index = all_index[n_train:]

    train_X = X_f[train_index, :] 
    train_Y = Y_f[train_index] 

    test_X = X_f[test_index, :]
    test_Y = Y_f[test_index] 

    print(train_X.shape, train_Y.shape)
    print(test_X.shape, test_Y.shape)
    
    # XGboost
    clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
                                 max_depth=1, random_state=0).fit(train_X, train_Y)
    # SVM
    clf_svm = SVC().fit(train_X, train_Y)
    
    # grid search
    param_grid = {'C': [0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001]}
    clf_grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2).fit(train_X, train_Y)
    
    #KNN  ##1NN
    clf_knn = KNeighborsClassifier(n_neighbors=1)
    clf_knn.fit(train_X, train_Y)
    
    #Decision tree
    clf_dtree = DecisionTreeClassifier()
    clf_dtree.fit(train_X, train_Y)
    
    #Random forest
    clf_rfc = RandomForestClassifier(n_estimators=100)
    clf_rfc.fit(train_X, train_Y)
    
    
    
    
    
    # Linear
    clf_LR = LinearRegression()
    clf_LR.fit(train_X, train_Y)
    
    # Logistic
    clf_LOG = LogisticRegression()
    clf_LOG.fit(train_X, train_Y)
    
    ##fm
    clf_fm = xl.FMModel(task='binary', lr=0.2, reg_lambda= 0.02)
    clf_fm.fit(train_X, train_Y, eval_set=[test_X, test_Y])
    #pred_y = fm_model.predict(test_X)
    #pred_y = pred_y.flatten()
    #y_pred = np.where(pred_y > 0.5, 1, 0)
    
    '''##ffm
    fields = [range(len(X_f))]
    ffm_model = xl.FFMModel(task='binary', lr=0.2, reg_lambda= 0.02)
    ffm_model.fit(train_X, train_Y, fields)
    pred_yy = ffm_model.predict(test_X)
    pred_yy = pred_yy.flatten()
    yy_pred = np.where(pred_yy > 0.5, 1, 0)'''
    
    
    
    
    
    print("****** XGboost: ")
    eval_metric(clf, test_X, test_Y)
    print("****** SVM:")
    eval_metric(clf_svm, test_X, test_Y)
    print("****** Grid search: " )
    eval_metric(clf_grid, test_X, test_Y)
    print("****** KNN(N=1): ")
    eval_metric(clf_knn, test_X, test_Y)
    print("****** Decision tree: ")
    eval_metric(clf_dtree, test_X, test_Y)
    print("****** Random forest: ")
    eval_metric(clf_rfc, test_X, test_Y)
    
    print("****** Linear: ")
    eval_mix(clf_LR, test_X, test_Y)
    print("****** Logistic: ")
    eval_mix(clf_LOG, test_X, test_Y)
    print("****** fm: ")
    eval_mix(clf_fm, test_X, test_Y)
    
    '''###ffm
    print("****** ffm: " )
    print(confusion_matrix(test_Y, yy_pred))
    print("F1 scores:", f1_score(test_Y, yy_pred, average='weighted')) 
    print("Report:", classification_report(test_Y, yy_pred))'''
  
   
    
   

    
```
